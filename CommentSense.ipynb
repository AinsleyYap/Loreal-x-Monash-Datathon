```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)        # remove URLs
    text = re.sub(r"@\w+", "", text)           # remove @mentions
    text = re.sub(r"#\w+", "", text)           # remove hashtags
    text = re.sub(r"[^a-z0-9\s]", " ", text)   # remove special chars
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["cleanText"] = df["textOriginal"].apply(clean_text)
```

```python
# relevance
brand_terms = {
    "loreal_brands": {
        "loreal", "l‚Äôoreal", "lorealparis", "loreal group", "loreal usa",
        # Luxury Division
        "lancome", "lancome paris", "kiehls", "kiehl's", "ysl", "ysl beauty", "yves saint laurent",
        "urban decay", "it cosmetics", "mugler", "valentino beauty", "prada beauty", "carita",
        "maison margiela", "atelier cologne", "takami", "shu uemura",
        # Consumer Division
        "maybelline", "maybelline new york", "garnier", "nyx", "nyx professional makeup", 
        # Professional Division
        "kerastase", "redken", "loreal professional", "matrix", "pureology", "pulpriot",
        # Active Cosmetics Division
        "la roche-posay", "vichy", "skinceuticals", "cerave", "decleor",
        # Other notable brands
        "biotherm", "essie", "ralph lauren fragrance", "cacharel", "viktor&rolf", "diesel fragrance", 
        "paris", "helena rubinstein"
    },
    "makeup": {
        "makeup", "cosmetics", "beauty",
        "lipstick", "lip stick", "lippy", "foundation", "blush", "concealer", "mascara", "eyeliner", "eye liner",
        "primer", "powder", "bronzer", "highlighter", "palette", "lipgloss", "lip gloss", "lipliner", "lip liner",
        "brow", "brows", "eyebrow", "eyebrows", "eyeshadow", "eye shadow", "contour", "contouring", "setting", "setting spray",
        "tint", "liner", "make up", "cc cream", "bb cream", "color corrector", "kajal", "makeup remover", "make-up"
    },
    "skincare": {
        "skincare", "skin care", "dermocosmetics", "dermo-cosmetics",
        "serum", "cleanser", "cleansing", "toner", "retinol", "hyaluronic", "hyaluronic acid", "moisturizer", "moisturiser",
        "sunscreen", "sun screen", "sunblock", "spf", "cream", "lotion", "gel", "essence", "oil", "mask", "face mask",
        "sheetmask", "sheet mask", "scrub", "exfoliant", "exfoliator", "exfoliating", "collagen", "niacinamide",
        "vitamin c", "antiaging", "anti-aging", "antioxidant", "hydrating", "hydration", "brightening", "firming",
        "wrinkle", "dark spot", "acne", "blemish", "sensitive skin", "night cream", "eye cream", "treatment"
    },
    "fragrance": {
        "perfume", "fragrance", "cologne", "toilette", "parfum", "eau", "mist", "scent", "body spray",
        "aroma", "spray", "floral", "woody", "musk", "musky", "citrus", "fresh", "oriental", "amber", "vanilla",
        "eau de parfum", "eau de toilette", "edp", "edt", "perfumery", "eau de cologne"
    },
    "haircare": {
        "haircare", "hair care", "hair",
        "shampoo", "conditioner", "hairmask", "hair mask", "deep conditioner", "hairspray", "hair spray", "styling",
        "styling product", "keratin", "hair serum", "dye", "hair dye", "color", "haircolor", "hair color", "coloring",
        "balayage", "bleach", "bleaching", "hair toner", "haircream", "hair cream", "scalp", "scalp care",
        "scalp treatment", "hair oil", "hair loss", "thinning hair", "volumizing", "smoothing", "curls", "curly hair",
        "hair treatment", "leave-in", "dry shampoo"
    }
}

campaign_terms = {
    "revitalift", "revitalift laser", "glycolic", "glycolic acid", "vitamin", "vitamin c", "glow", "glow booster",
    "hyaluron", "hyaluron acid", "hyaluronic", "ageperfect", "age perfect", "true match", "truematch", "infallible",
    "telescopic", "skin genius", "youth code", "elvive", "men expert", "micellar", "micellar water", "color riche",
    "derma", "derma science", "pro-retinol", "everpure", "everstrong", "fructis", "air dry", "root cover up",
    "dream lengths", "superliner", "lash paradise", "voluminous", "rouge signature", "blow dry", "perfect¬†match"
}


# -------------------------
# 2. Tokenize helper
# -------------------------
def tokenize(text):
    return set(re.findall(r"[a-z0-9]+", str(text).lower()))

# -------------------------
# 3. Jaccard similarity
# -------------------------
def jaccard(a, b):
    if not a or not b:
        return 0.0
    return len(a & b) / len(a | b)

# -------------------------
# 4. Relevance scoring function
# -------------------------
def relevance_score(text):
    toks = tokenize(text)
    if not toks:
        return 0.0
    # Base term pool = all brand/product/campaign terms
    base_terms = set().union(*brand_terms.values()) | campaign_terms
    base_score = jaccard(toks, base_terms)
    # Best match across categories
    cat_score = max((jaccard(toks, v) for v in brand_terms.values()), default=0.0)
    return 0.4*base_score + 0.6*cat_score

# -------------------------
# 5. Apply to dataset
# -------------------------
subset["relevance"] = subset["cleanText"].apply(relevance_score)

# -------------------------
# 6. Quick sanity check
# -------------------------
print(subset[["cleanText", "relevance"]].head(10))
print(subset["relevance"].describe())
```

```python
# Category tagging using your existing brand_terms/tokenize helpers
def tag_category(text):
    toks = tokenize(text)
    if not toks:
        return "other"
    # score overlap with each category
    scores = {cat: len(toks & terms) for cat, terms in brand_terms.items()}
    best_cat, best_hits = max(scores.items(), key=lambda x: x[1])
    return best_cat if best_hits > 0 else "other"

subset["category"] = subset["cleanText"].apply(tag_category)
subset["category"].value_counts()
```

```python
import re
import pandas as pd
from urllib.parse import urlparse

# ---------- Patterns ----------
url_pat         = re.compile(r"(https?://\S+|www\.\S+)", re.IGNORECASE)
email_pat       = re.compile(r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,}", re.IGNORECASE)
phone_pat       = re.compile(r"(\+?\d[\d\-\s]{7,}\d)")  # broad phone / WhatsApp numbers
wa_pat          = re.compile(r"(whatsapp|wa\.me)", re.IGNORECASE)
tg_pat          = re.compile(r"(t\.me|telegram)", re.IGNORECASE)
repeats_char    = re.compile(r"(.)\1{3,}")             # !!!!, loooove
repeats_word    = re.compile(r"\b(\w+)\b(?:\W+\1\b){2,}", re.IGNORECASE)  # word x3+
emoji_pat       = re.compile(r"[\U0001F300-\U0001FAFF]")  # broad emoji block
promo_words     = {"free","giveaway","promo","discount","subscribe","follow",
                   "dm","inbox","link","code","coupon","deal","sale",
                   "limited","offer","click","buy","shop","shipping",
                   "bitcoin","crypto","btc","eth","usdt"}
cta_phrases     = {"check my","check out","click here","buy now","order now",
                   "link in bio","visit my","dm me","message me","join group"}

# Suspicious TLDs; tune to your dataset
suspicious_tlds = {".ru",".tk",".top",".xyz",".icu",".pw",".cn",".gq",".ga",".ml"}
# Allow-list: official/likely-safe domains to de-penalize
domain_whitelist = {
    "youtube.com","youtu.be","instagram.com","tiktok.com","facebook.com","fb.com",
    "twitter.com","x.com","loreal.com","lorealparisusa.com","loreal-paris.co.uk",
    "lancome-usa.com","maybelline.com","garnierusa.com","nyxcosmetics.com",
    "laroche-posay.us","kiehls.com","sephora.com","ulta.com","shopee","lazada"
}

def extract_domains(text: str):
    """Return set of domains in text, lowercased (without subpaths)."""
    domains = set()
    for m in url_pat.findall(text or ""):
        u = m if m.startswith("http") else f"http://{m}"
        try:
            dom = urlparse(u).netloc.lower()
            if dom.startswith("www."): dom = dom[4:]
            domains.add(dom)
        except Exception:
            pass
    return domains

def non_ascii_ratio(s: str) -> float:
    s = s or ""
    if not s.strip(): return 0.0
    na = sum(1 for ch in s if ord(ch) > 127)
    return na / max(len(s), 1)

def uppercase_ratio(s: str) -> float:
    s = s or ""
    letters = [c for c in s if c.isalpha()]
    if not letters: return 0.0
    ups = sum(1 for c in letters if c.isupper())
    return ups / len(letters)

def token_set(s: str):
    return set(re.findall(r"[a-z0-9]+", (s or "").lower()))

def spam_signals(text: str):
    """Return dict of signal_name -> (triggered(bool), weight, detail)."""
    t = str(text or "")
    toks = token_set(t)
    domains = extract_domains(t)

    # Base features
    has_url        = bool(domains)
    has_susp_tld   = any(any(dom.endswith(tld) for tld in suspicious_tlds) for dom in domains)
    has_email      = bool(email_pat.search(t))
    has_phone      = bool(phone_pat.search(t)) or bool(wa_pat.search(t))
    has_tg         = bool(tg_pat.search(t))
    has_repeatchar = bool(repeats_char.search(t))
    has_repeatword = bool(repeats_word.search(t))
    very_short     = len(t.strip()) < 3
    low_diversity  = (len(set(t.strip())) / max(len(t.strip()),1)) < 0.2
    em_count       = len(emoji_pat.findall(t))
    many_emojis    = em_count >= 6
    nonascii_high  = non_ascii_ratio(t) > 0.5
    caps_shouting  = uppercase_ratio(t) > 0.6 and len(t) >= 8
    many_mentions  = t.count("@") >= 3
    many_hashtags  = t.count("#") >= 5

    promo_hit      = len(toks & promo_words) > 0
    cta_hit        = any(phrase in t.lower() for phrase in cta_phrases)

    # Weights (tuneable)
    W = {
        "url": 0.8,
        "susp_tld": 1.0,
        "email": 0.8,
        "phone_wa": 0.7,
        "telegram": 0.7,
        "repeatchar": 0.4,
        "repeatword": 0.6,
        "very_short": 0.6,
        "low_diversity": 0.6,
        "many_emojis": 0.5,
        "nonascii_high": 0.6,
        "caps_shouting": 0.5,
        "many_mentions": 0.5,
        "many_hashtags": 0.5,
        "promo_hit": 0.9,
        "cta_hit": 0.9,
    }

    # Domain whitelist de-weights the 'url' signal
    url_weight = 0.0 if any(dom in domain_whitelist for dom in domains) else W["url"]

    signals = {
        "url":            (has_url,        url_weight,        f"domains={list(domains)}" if domains else ""),
        "susp_tld":       (has_susp_tld,   W["susp_tld"],     ""),
        "email":          (has_email,      W["email"],        ""),
        "phone_wa":       (has_phone,      W["phone_wa"],     ""),
        "telegram":       (has_tg,         W["telegram"],     ""),
        "repeatchar":     (has_repeatchar, W["repeatchar"],   ""),
        "repeatword":     (has_repeatword, W["repeatword"],   ""),
        "very_short":     (very_short,     W["very_short"],   ""),
        "low_diversity":  (low_diversity,  W["low_diversity"],""),
        "many_emojis":    (many_emojis,    W["many_emojis"],  f"count={em_count}" if em_count else ""),
        "nonascii_high":  (nonascii_high,  W["nonascii_high"],""),
        "caps_shouting":  (caps_shouting,  W["caps_shouting"],""),
        "many_mentions":  (many_mentions,  W["many_mentions"],f"@={t.count('@')}"),
        "many_hashtags":  (many_hashtags,  W["many_hashtags"],f"#={t.count('#')}"),
        "promo_hit":      (promo_hit,      W["promo_hit"],    f"hit={list(toks & promo_words)}" if promo_hit else ""),
        "cta_hit":        (cta_hit,        W["cta_hit"],      ""),
    }
    return signals

def spam_score_and_reason(text: str, threshold=1.6):
    sigs = spam_signals(text)
    score = sum(w for (on,w,_) in sigs.values() if on)
    # Require at least two independent signals unless a single very-strong (>=1.0) signal fires
    strong = any(on and w >= 1.0 for (on,w,_) in sigs.values())
    num_on = sum(1 for (on,_,_) in sigs.values() if on)
    is_spam = (score >= threshold and num_on >= 2) or (strong and num_on >= 1)
    # Build reason string
    reasons = [f"{k}({w}){('['+d+']') if d else ''}" for k,(on,w,d) in sigs.items() if on]
    return is_spam, round(score,2), "; ".join(reasons)

# Apply to your subset
subset[["is_spam","spam_score","spam_reason"]] = subset["cleanText"].apply(
    lambda x: pd.Series(spam_score_and_reason(x))
)

# (Optional) If you prefer 0/1 like before:
subset["spam"] = subset["is_spam"].astype(int)

subset[["cleanText","spam","spam_score","spam_reason"]].head(60)
print("Spam rate:", subset["spam"].mean())
```

```python
subset["likeCount"] = pd.to_numeric(subset["likeCount"], errors="coerce").fillna(0)

subset["like_norm"] = subset.groupby("videoId")["likeCount"].transform(
    lambda s: (s - s.min()) / (s.max() - s.min() + 1e-6)
)

```

```python
subset["quality_score"] = (
    0.4*subset["relevance"] +
    0.01*subset["sentiment_score_1to5"] +
    0.2*subset["like_norm"] -
    0.3*subset["spam"]
).clip(0, 1)

subset["is_quality"] = (subset["quality_score"] >= 0.6).astype(int)

video_kpis = subset.groupby("videoId").agg(
    total_comments=("commentId","count"),
    quality_comments=("is_quality","sum"),
    quality_ratio=("is_quality","mean"),
    avg_sentiment=("sentiment_score_1to5","mean"),
    spam_rate=("spam","mean"),
    avg_relevance=("relevance","mean")
).reset_index().sort_values("quality_ratio", ascending=False)
```
